```python
soil_labels = {
    'Black Soil': 0,
    'Red soil': 1,
    'Clay soil': 2,
    'Alluvial soil': 3
}

harvest_map = {
    'Black Soil': 1,
    'Red soil': 1,
    'Clay soil': 0,
    'Alluvial soil': 1
}

fertility_tips = {
    'Black Soil': 'Add organic matter to enhance fertility.',
    'Red soil': 'Use compost and green manure to improve fertility.',
    'Clay soil': 'Improve drainage and add organic compost.',
    'Alluvial soil': 'Apply balanced fertilization to sustain fertility.'
}
```

```python
data_dir = "soil_dataset/Dataset/Train"
```

```python
import os
import cv2
import numpy as np
from tensorflow.keras.utils import to_categorical

images = []
texture_labels = []
harvest_labels = []

for label in soil_labels:
    folder_path = os.path.join(data_dir, label)
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)
        img = cv2.imread(img_path)
        if img is None:
            continue  # Skip unreadable images
        img = cv2.resize(img, (128, 128))
        images.append(img)
        texture_labels.append(soil_labels[label])
        harvest_labels.append(harvest_map[label])

images = np.array(images) / 255.0
texture_labels = to_categorical(texture_labels, num_classes=4)
harvest_labels = np.array(harvest_labels)
```

```python
from sklearn.model_selection import train_test_split

# Split into training and validation
X_train, X_val, y_train, y_val = train_test_split(images, texture_labels, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2, 2),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    Flatten(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dense(4, activation='softmax')  # 4 soil types
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=32)
```

```python
model.save("soil_texture_classifier.h5")
```

```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
import cv2
from google.colab import files

# Load the trained model
model = load_model("soil_texture_classifier.h5")

# Upload image
uploaded = files.upload()

# Soil classes used during training
soil_labels = {
    'Black Soil': 0,
    'Red soil': 1,
    'Clay soil': 2,
    'Alluvial soil': 3
}

# Mapping for harvesting suitability
harvest_map = {
    'Black Soil': 1,
    'Red soil': 1,
    'Clay soil': 0,
    'Alluvial soil': 1
}

# Fertility tips for each soil type
fertility_tips = {
    'Black Soil': 'Add organic matter to enhance fertility.',
    'Red soil': 'Use compost and green manure to improve fertility.',
    'Clay soil': 'Improve drainage and add organic compost.',
    'Alluvial soil': 'Apply balanced fertilization to sustain fertility.'
}

# Reverse mapping from index to soil label
label_map = {v: k for k, v in soil_labels.items()}

# Set threshold for confidence — predictions below this are rejected
CONFIDENCE_THRESHOLD = 0.5

# Process uploaded images
for img_name in uploaded:
    img = cv2.imread(img_name)
    if img is None:
        print(f"❌ Could not read image: {img_name}")
        continue

    # Resize and normalize the image
    img_resized = cv2.resize(img, (128, 128))
    img_input = np.expand_dims(img_resized / 255.0, axis=0)

    # Predict using model
    pred = model.predict(img_input)
    confidence = np.max(pred)
    predicted_label_index = np.argmax(pred)

    if confidence < CONFIDENCE_THRESHOLD:
        print(f"🚫 Not a valid soil image. (Prediction confidence: {confidence:.2f})")
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(f"Invalid Soil Image (Confidence: {confidence:.2f})")
        plt.axis('off')
        plt.show()
    else:
        predicted_label = label_map[predicted_label_index]
        print(f"🌱 Predicted Soil Texture: {predicted_label}")
        print(f"🧪 Harvesting Suitability: {'✅ Suitable for harvesting' if harvest_map[predicted_label] else '❌ Not suitable for harvesting'}")
        print(f"💡 Fertility Tip: {fertility_tips[predicted_label]}")
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(f"{predicted_label} (Confidence: {confidence:.2f})")
        plt.axis('off')
        plt.show()
```
